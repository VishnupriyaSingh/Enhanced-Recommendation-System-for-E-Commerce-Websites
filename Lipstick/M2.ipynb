```python
my_file = open("Text Reviews.txt", "r")
  
# reading the file
data = my_file.read()
  
# replacing end splitting the text 
# when newline ('\n') is seen.
documents = data.split("\n")
my_file.close()
```

```python
lower_case_documents = []
for i in documents:
    lower_case_documents.append(i.lower())
print(lower_case_documents)
```

```python
sans_punctuation_documents = []
import string

for i in lower_case_documents:
    sans_punctuation_documents.append(''.join(c for c in i if c not in string.punctuation))
    
print(sans_punctuation_documents)
```

```python
from nltk.stem import PorterStemmer
 
st = PorterStemmer()
 
output = []
for sentence in sans_punctuation_documents:
    output.append(" ".join([st.stem(i) for i in sentence.split()]))
 
for item in output:
    print(item)
```

```python
preprocessed_documents = []
for i in output:
    preprocessed_documents.append(i.split(' '))
print(preprocessed_documents)
```

```python
frequency_list = []
import pprint
from collections import Counter

for i in preprocessed_documents:
    frequency_list.append(Counter(i))
    
pprint.pprint(frequency_list)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer(stop_words='english')
count_vector.fit(documents)
count_vector.get_feature_names_out()
```

```python
doc_array = count_vector.transform(documents).toarray()
doc_array
```

```python
frequency_matrix = pd.DataFrame(doc_array,index=documents,columns=count_vector.get_feature_names_out())
frequency_matrix
```
